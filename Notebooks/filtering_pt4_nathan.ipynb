```python
# Define the split_location function
def split_location(location):
    location = str(location)
    if location.upper() == 'WASHINGTON, D.C.':
        return ['Washington, D.C.', '']
    else:
        parts = location.split(', ')
        city = parts[0]
        region = ', '.join(parts[1:]) if len(parts) > 1 else ''
        return [city, region]

# Define the capitalize_words function
def capitalize_words(string):
    return ' '.join(word.capitalize() for word in string.split())

# Split 'Location' into 'City' and 'Region'
df[['City', 'Region']] = df['Location'].apply(split_location).apply(pd.Series)

# Capitalize 'City' and 'Region'
df['City'] = df['City'].apply(capitalize_words)
df['Region'] = df['Region'].apply(capitalize_words)

# Replace state abbreviations with full names
us_states = {
    # (The previously defined dictionary of state abbreviations and full names)
}

df['Region'] = df['Region'].apply(lambda x: us_states.get(x.upper(), x))

# Display the first few rows to verify the changes
print(df[['Location', 'City', 'Region']].head(50))
```
