```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plane_data_df = pd.read_csv('Plane_Data/Aviation_Data.csv')
plane_data_df.head(50)
```

```python
plane_data_df['People.Count'] = plane_data_df['Total.Fatal.Injuries'] + plane_data_df['Total.Serious.Injuries'] + plane_data_df['Total.Minor.Injuries'] + plane_data_df['Total.Uninjured']
plane_data_df
```

```python
#df = df.dropna()
plane_data_df = plane_data_df.fillna(0)
plane_data_df.head(10)
```

```python
filtered_plane_data_df = plane_data_df[(plane_data_df['Number.of.Engines'] <= 2) & (plane_data_df['People.Count'] <= 20)]
filtered_plane_data_df
```
